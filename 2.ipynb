```python
import os
import cv2
```

```python
img_path = os.path.join('img.png')
```

```python
img = cv2.imread(img_path)
```

```python
cv2.imshow('imgage',img)
cv2.waitKey(0)
```

```python
webcam = cv2.VideoCapture(0)
```

```python
while True :
    ret , frame = webcam.read()
    cv2.imshow('frame',frame)
    if cv2.waitKey(40) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```
