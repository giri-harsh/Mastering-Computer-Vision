# Implimenting Canny Edge detection


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
```

```python
face = cv2.imread(r"C:\Work\Programing Language\Python\Libraries\OpenCV\data\img_3.png")
plt.imshow(face)
```

```python
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
edge = cv2.Canny(face,threshold1=127,threshold2=127)
plt.imshow(edge)
```

## Formula to improve `CANNY EDGE DETECTION`

```python
med_val = np.median(img)
med_val
```

```python
# set threshold val to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# set either 130% of median or the max value of 255 which ever is smaller
upper =int(min(255,1.3*med_val))
```

* `LOWER` - set threshold val to either 0 or 70% of the median value whichever is greater
* `UPPER` - set either 130% of median or the max value of 255 which ever is smaller

```python
edge = cv2.Canny(img,threshold1=lower,threshold2=upper)
plt.imshow(edge)
```

* `OBSERVATION` - edges in image is getting worse

```python
blured = cv2.blur(img,ksize=(10,10))
blured = cv2.cvtColor(blured,cv2.COLOR_BGR2RGB)
plt.imshow(blured)
```

```python
edge = cv2.Canny(blured,threshold1=lower,threshold2=upper)
plt.imshow(edge)
```
