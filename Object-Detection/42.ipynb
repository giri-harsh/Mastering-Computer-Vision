```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

 ## This will basically show the heatmap where the it has the higher values where it thinks it have higher corelations

```python
for m in methods :
     # create a copy of img
    full_copy = full.copy()
    method = eval(m)

     # template matching
    result = cv2.matchTemplate(full_copy,face,method)
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(result)
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else :
        top_left = max_loc

    height,width,channel = face.shape
    bottom_right = (top_left[0]+width , top_left[1]+height)
    cv2.rectangle(full_copy,pt1=top_left,pt2=bottom_right,color=(0,0,255),thickness=10)


#      plot and show
     #      img
    plt.subplot(121)
    plt.imshow(result)
    plt.title("Heatmap")

    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("Detection of template")
    plt.suptitle(m)
    plt.show()







    plt.subplot(121)

```
