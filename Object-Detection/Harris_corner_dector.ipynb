# for corner detection we need float values in place of int
# this is for `harris corner detection`

```python
gray = np.float32(gray_chess)
```

```python
dst = cv2.cornerHarris(src = gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst.shape
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01 *dst.max()] = [255,0,0] #set it to rgb for better visualiszation
```

```python
plt.imshow(flat_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
gray_real_chess = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src=gray_real_chess,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()] = [255,0,0]
plt.imshow(real_chess)
```
