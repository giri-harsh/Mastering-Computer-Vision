```python
import os
import cv2
import numpy as np

```

## OpenCV Events

```python
img = np.zeros((512,512,3))
```

```python
def draw(event , x ,y, flags, params):
    print(event)

cv2.namedWindow(winname='window')
cv2.setMouseCallback('window',draw)
while True:
    cv2.imshow('window',img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```
