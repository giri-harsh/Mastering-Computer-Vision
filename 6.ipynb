```python
import os
import cv2
import numpy as np

```

## Creating Shapes on Images

```python
img = np.zeros((512,512,3))
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(255,0,0),thickness=-1)
cv2.imshow('window',img)
cv2.waitKey(0)
```

```python
cv2.circle(img,center=(100,400),radius=50 , color= (0,0,255),thickness=3)
cv2.imshow('window',img)
cv2.waitKey(0)
```

```python
cv2.line(img,pt1 = (0,0),pt2=(512,512),color=(0,255,0),thickness=3)
cv2.imshow('window',img)
cv2.waitKey(0)
```

```python
cv2.putText(img,org = (100,100),fontScale=2,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text = 'Hello World',fontFace=cv2.FONT_HERSHEY_SIMPLEX)
cv2.imshow('window',img)
cv2.waitKey(0)
```
